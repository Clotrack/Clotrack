## ¿Qué es Jupyter?
- Es un entorno donde mezclas **código** + **texto bonito** + **imágenes** + **resultados** en el mismo sitio.

````python
# En Colab a veces hay que forzar la versión buena de ipywidgets
!pip install ipywidgets --upgrade -q
import ipywidgets as widgets
from IPython.display import display, clear_output
print("¡Widgets listos! 🎛️")
````

Slider Básico

````python
slider = widgets.IntSlider(
    value=42,
    min=0,
    max=100,
    step=1,
    description='Número:',
    style={'description_width': 'initial'},
    continuous_update=False
)

def on_value_change(change):
    print(f"Nuevo valor: {change['new']}")

slider.observe(on_value_change, names='value')
display(slider)
````

Slider que mueve una gráfica en tiempo real

````python
import matplotlib.pyplot as plt
import numpy as np

# Widget
frecuencia = widgets.FloatSlider(value=1, min=0.1, max=5, step=0.1, description='Frecuencia:')

# Función que se llama cada vez que mueves el slider
def actualizar_grafica(change):
    clear_output(wait=True)
    display(frecuencia)

    x = np.linspace(0, 2*np.pi, 500)
    y = np.sin(frecuencia.value * x)

    plt.figure(figsize=(10,4))
    plt.plot(x, y, lw=3, color="#ff6f61")
    plt.title(f"Onda senoidal con frecuencia = {frecuencia.value:.2f}", fontsize=16)
    plt.grid(True, alpha=0.3)
    plt.ylim(-1.1, 1.1)
    plt.show()

frecuencia.observe(actualizar_grafica, names='value')
display(frecuencia)
actualizar_grafica(None)  # primera ejecución
````

Varios widgets juntos

````python
# Widgets
nombre = widgets.Text(value="Ada Lovelace", description="Nombre:")
edad = widgets.IntSlider(value=25, min=0, max=120, description="Edad:")
color = widgets.ColorPicker(value='#ff9f1c', description="Color favorito:")
boton = widgets.Button(description="¡Saludar!", button_style='success', icon='check')

# Output
out = widgets.Output()

def saludar(b):
    with out:
        clear_output()
        print(f"¡Hola {nombre.value}! 🎉")
        print(f"Tienes {edad.value} años")
        print(f"Tu color favorito es {color.value}")
        print(" " * 10 + "✨" * edad.value)

boton.on_click(saludar)

# Mostrar todo bonito
display(nombre, edad, color, boton, out)
````

Dropdown + Checkbox + Play (animación)

````python
# Widget de reproducción automática
play = widgets.Play(value=50, min=0, max=100, step=1, interval=100, description="Animar:")

# Slider conectado al play
tiempo = widgets.FloatSlider(min=0, max=10, step=0.1, description="Tiempo:")

# Conectar play y slider
widgets.jslink((play, 'value'), (tiempo, 'value'))

# Función de animación
def animar(t):
    clear_output(wait=True)
    display(widgets.HBox([play, tiempo]))

    x = np.linspace(0, 2*np.pi, 400)
    y = np.sin(2*np.pi*t/10 * x) * np.exp(-0.1*t)

    plt.figure(figsize=(10,4))
    plt.plot(x, y, color="#7209b7", lw=4)
    plt.title(f"Onda amortiguada (t = {t:.1f}s)", fontsize=18)
    plt.ylim(-1.1, 1.1)
    plt.grid(alpha=0.3)
    plt.show()

tiempo.observe(lambda change: animar(change['new']), names='value')
display(widgets.HBox([play, tiempo]))
animar(0)
````
